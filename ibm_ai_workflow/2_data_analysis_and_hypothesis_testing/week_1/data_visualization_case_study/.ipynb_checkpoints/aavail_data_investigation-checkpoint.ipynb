```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
from aavail_data_ingestor import Aavail_Data_Ingestor as ADI
```

```python
columns = 'customer.last_name, \
    customer.first_name, \
    customer.gender, \
    customer.DOB, \
    customer.city, \
    customer.state, \
    customer.country_id, \
    customer.customer_id, \
    invoice.signup_date, \
    invoice.last_stream'
connection_path = './aavail-customers.db'
df_streams_path = './aavail-streams.csv'
pd_query = 'SELECT {} FROM customer JOIN invoice ON customer.customer_id = invoice.customer_id;'.format(columns)
```

```python
data_ingestor = ADI(connection_path, df_streams_path, pd_query)
```

```python
processed_customer_data, processed_stream_data = data_ingestor.begin_cleaning_pipeline(verbose = True)
```

```python
processed_customer_data
```

```python
processed_customer_data.isnull().sum().sum()
```

```python
processed_stream_data.isnull().sum().sum()
```

```python
processed_customer_data['country_id'] = processed_customer_data['country_id'].map({0: 'singapore', 1: 'united_states'})
```

```python
df = processed_customer_data.drop(['state'], axis = 1)
```

```python
df.describe()
```

```python
sub_type_mapping_data = pd.read_sql_query('SELECT * from invoice_item', sqlite3.connect(connection_path))
sub_type_mapping_data
stmd = dict(sub_type_mapping_data[['invoice_item_id', 'invoice_item']])
stmd
```

```python
stmd = dict(stmd['invoice_item'])
sub_type_map = dict()
for key in stmd:
    sub_type_map[key + 1] = stmd[key]
sub_type_map, type(sub_type_map)
```

```python
df['sub_type'] = df['sub_type'].map(sub_type_map)
df
```

```python
sns.catplot(x = 'country_id', 
            y = 'length_of_activity', 
            hue = 'gender', 
            split = True, 
            col = 'is_sub', 
            kind = 'violin', 
            bw = .15, 
            cut = 0,
            data = df)
```

### Showcasing the length of activity, per country, per gender, starting from initial use of the product, to the last time the product was used.

Distribution of the length of activity between subscibers and non-subscribers in their repective nations are roughly the same. The most notable difference is in Singapores women, which are more likely to positively stray farther from the typical watch time, although the number of people doing so is not large.

```python
g = sns.catplot(x = 'country_id', hue = 'gender', col = 'is_sub', kind = 'count', data = df)
```

There are a much larger portion of non-subscribers in the United States, in comparison to singapore.
Non-subscribers in the US are predominantly male.
There are much more subscibers in Singapore in comparison to non-subscibers.
However, far fewer people ever use the platform in Singapore.
This only makes sense as the population of Singapore is far less, and the size of the nation is far smaller as well.

```python
time_sorted_df = df.sort_values(by = 'signup_date')
time_sorted_df['count'] = range(1, 1001)
time_sorted_df
```

```python
g = sns.relplot(x = 'signup_date', y = 'count', kind = 'line', data = time_sorted_df)
g.fig.autofmt_xdate()
```

Over time the amount of subscribers increases almost linearly.

```python
g = sns.relplot(x = 'signup_date', y = 'count', kind = 'line', data = time_sorted_df.loc[time_sorted_df['country_id'] == 'singapore'])
g.fig.autofmt_xdate()
```

```python
g = sns.relplot(x = 'signup_date', y = 'count', kind = 'line', data = time_sorted_df.loc[time_sorted_df['country_id'] == 'united_states'])
g.fig.autofmt_xdate()
```

```python
reverse_count_np = time_sorted_df['count'].to_numpy()[::-1]
is_sub_np = time_sorted_df['is_sub'].to_numpy()

for i in range(len(is_sub_np)):
    if is_sub_np[i] == 0:
        reverse_count_np[i:] -= 1

time_sorted_df['reversed_count'] = reverse_count_np
```

```python
g = sns.relplot(x = 'signup_date', y = 'reversed_count', kind = 'line', data = time_sorted_df.loc[time_sorted_df['country_id'] == 'singapore'])
g.fig.autofmt_xdate()
```

```python
g = sns.relplot(x = 'signup_date', y = 'reversed_count', kind = 'line', data = time_sorted_df.loc[time_sorted_df['country_id'] == 'united_states'])
g.fig.autofmt_xdate()
```

Overall the decrease in subscribers as well as the increase in subscribers, have a fairly linear increase/decrease.

```python
g = sns.relplot(x = 'signup_date', y = 'reversed_count', kind = 'line', data = time_sorted_df)
g.fig.autofmt_xdate()
```
